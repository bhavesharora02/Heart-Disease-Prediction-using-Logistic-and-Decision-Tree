Importing the Dependencies!

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

Data Collection and Processing

```python
#loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart_disease_data.csv')
```

```python
#print first five rows of the dataset
heart_data.head()
```

```python
#print last five rows of the dataset
heart_data.tail()
```

```python
heart_data.shape
```

```python
#getting some info about the data
heart_data.info()
```

```python
 #statistics measures about the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data['target'].value_counts()
```

1 --> Defective Heart

0 --> Healthy Heart

Splitting the Features and Target

```python
X = heart_data.drop(columns='target',axis=1) 
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression Model - used for Classfication

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X_train, Y_train)
```

```python
reg = LinearRegression().fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_prediction_logistics = model.predict(X_train)
training_data_accuracy_logistics = accuracy_score(X_train_prediction_logistics, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy_logistics)
```

```python
X_train_prediction_dt = clf.predict(X_train)
training_data_accuracy_dt = accuracy_score(X_train_prediction_dt, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy_dt)
```

```python
#accuracy on test data
X_test_prediction_logistics = model.predict(X_test)
training_data_accuracy_logistics = accuracy_score(X_test_prediction_logistics, Y_test)
```

```python
print('Accuracy on Test Data : ', training_data_accuracy_logistics)
```

```python
X_test_prediction_dt = clf.predict(X_test)
test_data_accuracy_dt = accuracy_score(X_test_prediction_dt, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy_dt)
```

Building a Predictive System

```python
input_data = (57,0,0,120,354,0,1,163,1,0.6,2,0,2)

#change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

#reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==0):
    print('The Person does not have heart disease')
else:
    print('The Person has heart disease')
```
